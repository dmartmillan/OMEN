```python
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
# Quantification of different link types in different alpha values

top_genes = [50, 80, 100, 120, 150, 200]

result = {}

for alpha in [0.2,0.5,0.65,0.8,0.9]:

    # Check file name!
    with open('./output/alpha_' + str(alpha) + '/links_vs_N', 'r') as f:
        
        result[alpha] = {'cannot_link' : [], 'must_link': []}
        
        lines = f.read()
        for word in lines.split("\n"):
            if (word != ""):
                r1 = re.findall(r"Cannot links: [0-9]*",word)
                r2 = re.findall(r"[0-9]+",r1[0])
                result[alpha]['cannot_link'].append(eval(r2[0]))
                r3 = re.findall(r"Must links: [0-9]*",word)
                r4 = re.findall(r"[0-9]+",r3[0])
                result[alpha]['must_link'].append(eval(r4[0]))
            
print(result)
```

```python
for alpha in [0.2,0.5,0.65,0.8,0.9]:
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    links = result[alpha]['cannot_link']
    ax.bar(top_genes,links, width = 12)
    plt.title("Cannot links")
    plt.xlabel("Top genes")
    plt.ylabel("Number of links")
    plt.savefig("cannot_links_" + str(alpha) + ".png", bbox_inches='tight')
    plt.show()
    
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    links = result[alpha]['must_link']
    ax.bar(top_genes,links, width = 12)
    plt.title("Must links")
    plt.xlabel("Top genes")
    plt.ylabel("Number of links")
    plt.savefig("must_links_" + str(alpha) + ".png", bbox_inches='tight')
    plt.show()
```
