```python
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Evaluation of threshold distribution in a OMEN's run

thresholds_prostate = []
with open('./output/run_output_folder/output_file_1.evaluated') as f:
    for line in f:
        value = re.findall(r"[0-9].[0-9]+-",line)[0][:-1]
        thresholds_prostate.append(eval(value))
```

```python
thresholds_prostate
```

```python
thresholds_values_prostate = np.array(thresholds_prostate)

fig = plt.gcf()
plt.hist(thresholds_values_prostate, bins=100)
plt.title("Frequency Prostate cancer")
plt.xlabel("Threshold")
plt.ylabel("Frequency of Threshold")
plt.show()
plt.draw()
```
