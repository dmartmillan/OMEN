```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = [0.2,0.5,0.65,0.8,0.9]
TP53 = [3,2,2,3,3]
SPOP = [373,790,751,1886,2121]
TTN = [124,240,1030,1307,1518]
MUC16 = [562,np.nan,np.nan,np.nan,np.nan]
```

```python
fig, ax = plt.subplots()
ax.set_yscale('log')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.scatter(alpha, TP53,label='TP53')
plt.scatter(alpha, SPOP,label='SPOP')
plt.scatter(alpha, TTN,label='TTN')
plt.scatter(alpha, MUC16,label='MUC16')
for i, j in zip(alpha,TP53):
    plt.text(i, j+0.4, '{}'.format(j))
for i, j in zip(alpha,SPOP):
    plt.text(i, j+0.5, '{}'.format(j))
for i, j in zip(alpha,TTN):
    plt.text(i, j+0.5, '{}'.format(j))
for i, j in zip(alpha,MUC16):
    plt.text(i, j+0.5, '{}'.format(j))
leg = ax.legend();
# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('           Ranking of genes with a high mutation frequency in function of alpha',pad=20)
plt.ylabel('Ranking')
plt.xlabel('Alpha')
plt.savefig('alpha_vs_ranking_most_mutated_genes')
```
