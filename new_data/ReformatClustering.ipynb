```python
inNetworkFileName = 'graph'
outNetworkFileName = 'graph.adjacency'

outNodeIndexFileName = 'nodes.index'

inMustFileName = 'must'
outMustFileName = 'must.links'

inCannotFileName = 'cannot'
outCannotFileName = 'cannot.links'

nodeCount=0
edgeCount=0
nodeSet = {}
adjacency = []
with open(inNetworkFileName,'r') as networkFile:
    # skip header
    networkFile.readline()
    # process lines
    for line in networkFile.readlines():
        split = line.strip('\n\r').split(',')
        if not split[0] in nodeSet:
            nodeCount += 1
            nodeSet[split[0]] = nodeCount - 1
            adjacency.append([])
        if not split[1] in nodeSet:
            nodeCount += 1
            nodeSet[split[1]] = nodeCount - 1
            adjacency.append([])
        adjacency[nodeSet[split[0]]].append(nodeSet[split[1]])
        adjacency[nodeSet[split[1]]].append(nodeSet[split[0]])
        edgeCount += 1
        
with open(outNodeIndexFileName,'w') as nodeFile:
    for node in nodeSet:
        print(node + " " + str(nodeSet[node]), file = nodeFile)
        
with open(outNetworkFileName,'w') as networkFile:
    print(str(nodeCount) + " " + str(edgeCount), file = networkFile)
    for nodes in adjacency:
        print(" ".join(str(x) for x in nodes), file = networkFile)


def convertLinksFile(nodeCount, inLinksFileName, outLinksFileName):
    links = []
    for i in range(nodeCount):
        links.append([])
    constraintCount = 0
    with open(inLinksFileName,'r') as linksFile:
        # skip header
        linksFile.readline()
        # process lines
        for line in linksFile.readlines():
            split = line.strip('\n\r').split(',')
            links[nodeSet[split[0]]].append(nodeSet[split[1]])
            links[nodeSet[split[0]]].append(split[2])
            links[nodeSet[split[1]]].append(nodeSet[split[0]])
            links[nodeSet[split[1]]].append(split[2])
            constraintCount += 1
    with open(outLinksFileName,'w') as linksFile:
        print(str(nodeCount) + " " + str(constraintCount), file = linksFile)
        for linksFromNode in links:
            print(" ".join(str(x) for x in linksFromNode), file = linksFile)

convertLinksFile(nodeCount, inMustFileName, outMustFileName)
convertLinksFile(nodeCount, inCannotFileName, outCannotFileName)
```
