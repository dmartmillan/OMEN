```python
import pandas as pd
from  matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import textwrap
from matplotlib_venn import venn3
```

```python
rankings = {}
rankings_aggr = {}
rankings_non_aggr = {}
rankings_nw8 = {}
rankings_nw8_aggr = {}
rankings_nw8_non_aggr = {}

alpha = 0.65

N = 200
```

```python
df = pd.read_csv('ranks/alpha_'+str(alpha)+'.desc',header = None)
df = df[df.index<N]
rankings = df[0].tolist()

df = pd.read_csv('ranks/alpha_'+str(alpha)+'_aggr.desc',header = None)
df = df[df.index<N]
rankings_aggr = df[0].tolist()

df = pd.read_csv('ranks/alpha_'+str(alpha)+'_non_aggr_reduced.desc',header = None)
df = df[df.index<N]
rankings_non_aggr = df[0].tolist()
```

```python
df = pd.read_csv('ranks/network8_alpha_'+str(alpha)+'.desc',header = None)
df = df[df.index<N]
rankings_nw8 = df[0].tolist()

df = pd.read_csv('ranks/network8_alpha_'+str(alpha)+'_aggr.desc',header = None)
df = df[df.index<N]
rankings_nw8_aggr = df[0].tolist()

df = pd.read_csv('ranks/network8_alpha_'+str(alpha)+'_non_aggr_reduced.desc',header = None)
df = df[df.index<N]
rankings_nw8_non_aggr = df[0].tolist()
```

```python
def plot_Venn_diagram(ranking1,ranking2,ranking3,alpha,network_nr):
    name = 'network_'+str(network_nr)+'_'+ 'allvsaggrvsnonaggr' + '_' + str(alpha)
    set1 = set(ranking1)
    set2 = set(ranking2)
    set3 = set(ranking3)
    v=venn3([set1, set2, set3], ('all', 'aggr', 'nonaggr'))
    plt.title('Venn diagram of the all, aggr and reduced nonaggr top ranked genes \n with \u03B1 = '+ str(alpha) + ' in prostate cancers and network '+str(network_nr) + '; ' + str(N) + ' top genes',size=12)
    plt.savefig('aggr_nonaggr_comparation_plots/' + 'Venn_diagram_'+name+'_nodes_' +str(N)+'_reduced.png',bbox_inches="tight")
    plt.show()
```

```python
plot_Venn_diagram(rankings,rankings_aggr,rankings_non_aggr,alpha,1)
```

```python
plot_Venn_diagram(rankings_nw8,rankings_nw8_aggr,rankings_nw8_non_aggr,alpha,8)
```
